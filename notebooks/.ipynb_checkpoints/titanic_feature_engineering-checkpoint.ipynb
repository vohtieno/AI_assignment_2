```python
# 1. Creating Derived Features


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Load the dataset 
df = pd.read_csv('../data/train.csv')

# for passangers travelling as a family
# 1. FamilySize: Combined count of siblings/spouses and parents/children + yourself 
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

```

```python
# for the passangers travelling alone

df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
```

```python
# 3. Title Extraction: Pulling 'Mr', 'Mrs', 'Miss', etc., from the Name column 

# The 'r' before the quotes fixes the invalid escape sequence warning
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False) 
```

```python
# 4. Deck Extraction: Taking the first letter from the Cabin column (if not dropped) 

df['Deck'] = df['Cabin'].apply(lambda x: str(x)[0] if pd.notnull(x) else 'U')
```

```python
# 5. Age Groups: Categorizing Age into life stages 

bins = [0, 12, 19, 59, 100]
labels = ['Child', 'Teen', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
# 6. Fare per person: 
df['FarePerPerson'] = df['Fare'] / df['FamilySize']
```

```python
## CATEGORICAL ENCODING

# One-hot encode nominal features (Sex, Embarked, Title, Deck) 
df = pd.get_dummies(df, columns=['Sex', 'Embarked', 'Title', 'Deck'], drop_first=True)

# Ordinal encoding for Pclass is already numerical (1, 2, 3), 
# but we can ensure it is treated as an ordered category 
df['Pclass'] = df['Pclass'].astype(int)
```

```python
## FEATURE TRANSFORMATIONS

# Log transform skewed features to normalize distribution 
df['Fare_log'] = np.log1p(df['Fare'])
df['Age_log'] = np.log1p(df['Age'])

# Standardize/Scale features (StandardScaler makes mean=0, variance=1) 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Scaling numerical features for distance-based models
scale_cols = ['Age', 'Fare', 'FamilySize', 'FarePerPerson']
df[scale_cols] = scaler.fit_transform(df[scale_cols])
```

```python
# ploting the figures for visualization

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['Fare'], kde=True)
plt.title('Original Fare Distribution (Skewed)')

plt.subplot(1, 2, 2)
sns.histplot(df['Fare_log'], kde=True)
plt.title('Log Transformed Fare (Normalized)')

plt.show()
```

```python
# Saving the folder

df.to_csv('../data/train_cleaned.csv', index=False)
print("File 'train_cleaned.csv' has been saved successfully!")
```
