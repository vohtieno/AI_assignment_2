```python

import pandas as pd
import numpy as np

# Load the dataset (adjust the path if your notebook is in a different directory)
df = pd.read_csv('data/train.csv')

```

```python
# Handling missing values

# Identify missing values in all columns
print("Missing values before cleaning:")
print(df.isnull().sum())
```

```python

# Impute 'Age' with the median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Impute 'Embarked' with the mode
mode_embarked = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(mode_embarked)

# Drop the 'Cabin' column due to too many missing values
df = df.drop(columns=['Cabin'])

# Optional: Create an indicator column for 'Age' missingness before imputing
# df['Age_was_missing'] = df['Age'].isnull().astype(int)
```

```python
# Outlier Handling

def cap_outliers(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Cap the outliers
    dataframe[column] = np.where(dataframe[column] > upper_bound, upper_bound, dataframe[column])
    dataframe[column] = np.where(dataframe[column] < lower_bound, lower_bound, dataframe[column])
    
    return dataframe

# Cap outliers for Fare and Age
df = cap_outliers(df, 'Fare')
df = cap_outliers(df, 'Age')
```

```python
# Check Data Consistency

# Fix inconsistencies in the 'Sex' column by converting to lowercase and stripping whitespace
df['Sex'] = df['Sex'].str.lower().str.strip()

# Check to ensure only 'male' and 'female' exist
print("Unique values in Sex:", df['Sex'].unique())
```

```python
# Checking for and removing duplicates

# Check for duplicates
print(f"Duplicates found: {df.duplicated().sum()}")

# Remove duplicates if any exist
df = df.drop_duplicates()
```

```python
# verifying that the data is clean

print(df.isnull().sum())
```

```python
# Saving the clean data to the data folder


df.to_csv('../data/train_cleaned.csv', index=False)
print("Cleaned dataset saved successfully!")
```
