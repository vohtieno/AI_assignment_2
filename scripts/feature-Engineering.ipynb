```python
import pandas as pd
import numpy as np

# importing data

train_df = pd.read_csv('../data/train.csv')
```

```python
# Creating derived features


# Family size
# Formula: SibSp + Parch + 1 (the passenger themselves)
train_df['FamilySize'] = train_df['SibSp'] + train_df['Parch'] + 1
```

```python
# A lone passanger
train_df['IsAlone'] = train_df['FamilySize'].apply(lambda x: 1 if x == 1 else 0)
```

```python
# title extraction for names
train_df['IsAlone'] = train_df['FamilySize'].apply(lambda x: 1 if x == 1 else 0)
```

```python
# deck extraction

train_df['Deck'] = train_df['Cabin'].astype(str).apply(lambda x: x[0] if x != 'nan' and x != 'Unknown' else 'U')
```
