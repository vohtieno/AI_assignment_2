```python
import pandas as pd
import numpy as np

#Load the dataset
train_df = pd.read_csv('../data/train.csv')
```

```python
train_df.isnull().sum()
```

```python
# Calculate the median age
median_age = train_df['Age'].median()
print(median_age)
```

```python
# Fill missing values with that median
train_df['Age'] = train_df['Age'].fillna(median_age)
```

```python
# Find the most frequent starting port
mode_embarked = train_df['Embarked'].mode()[0]

# Fill missing values with the mode
train_df['Embarked'] = train_df['Embarked'].fillna(mode_embarked)
```

```python
# 1. Create an indicator column (1 if missing, 0 if not)
train_df['Has_Cabin'] = train_df['Cabin'].apply(lambda x: 0 if pd.isna(x) else 1)

# 2. Since Cabin has too many missing values to guess correctly, 
train_df['Cabin'] = train_df['Cabin'].fillna('Unknown')
```

```python
# checking if the changes are implemented
train_df.isnull().sum()
```

```python
# Handling Outliers
import seaborn as sns
import matplotlib.pyplot as plt

# Visualizing outliers in Fare
plt.figure(figsize=(10, 5))
sns.boxplot(x=train_df['Fare'])
plt.title('Outliers in Fare')
plt.show()
```

```python
# Calculate IQR for Fare
Q1 = train_df['Fare'].quantile(0.25)
Q3 = train_df['Fare'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

# Count how many outliers we have
outliers = train_df[(train_df['Fare'] < lower_bound) | (train_df['Fare'] > upper_bound)]
print(f"Number of outliers in Fare: {len(outliers)}")
```

```python
# Capping Fare at the Upper Bound
train_df['Fare'] = np.where(train_df['Fare'] > upper_bound, upper_bound, train_df['Fare'])

# Capping Age 
# Let's say we cap age at 70 if someone is 80+
train_df['Age'] = np.where(train_df['Age'] > 70, 70, train_df['Age'])

# Verify the change
print(f"New Max Fare: {train_df['Fare'].max()}")
```

```python
#DATA CONSISTENCY

# Check unique values
print(train_df['Sex'].unique())

# Clean them up to be consistent
train_df['Sex'] = train_df['Sex'].str.lower().str.strip()
```

```python
# Removing duplicates

train_df.drop_duplicates(inplace=True)
```

```python
# Saving the cleaned data
train_df.to_csv('../data/train_cleaned.csv', index=False)
```
